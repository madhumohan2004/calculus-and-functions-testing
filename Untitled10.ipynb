```python
import sympy as sp
from sympy import symbols, diff
import numpy as np

# Define the variable and the expression
x = symbols('x')
expression = input("Enter a mathematical expression in terms of x: ")

# Find the derivative
derivative = diff(expression, x)

# Print the result
print(f"The derivative of {expression} with respect to x is: {derivative}")


# Get the value of x for evaluation
x_value = float(input("Enter the value of x for evaluation: "))


# Evaluate the derivative at the specified value of x
result = derivative.subs(x, x_value)

# Print the result
print(f"The derivative of {expression} at x = {x_value} is: {result}")
arr=np.array([1,2,3])
second_derivative= diff(derivative,x)
print(f"second derivative is {second_derivative}")
a=second_derivative.subs(x,x_value)
b=sp.lambdify(x,second_derivative,"numpy")
result=b(arr)
print(f"the value of he second derivative at the specified vlue of x {a}")
print(f"the value at each point {result}")
```
